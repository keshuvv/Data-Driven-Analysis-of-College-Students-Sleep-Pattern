# Research Question

Which measurable factors, reported through self-surveys—such as sleep duration, physical activity, and screen time—are most strongly associated with getting adequate sleep schedules (7–9 hours per night) among college students (ages 18-25) in the US?

## Background and Prior Work

Maintaining a healthy sleep schedule is essential for college students in the U.S, as it directly impacts their academic performance, mental health, and overall well-being. However, research from the Centers for Disease Control and Prevention (CDC) shows that around 75% of college students report experiencing sleep disturbances or poor sleep quality, largely due to demanding schedules and the challenges of balancing coursework, social activities, and work commitments.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Lifestyle factors such as caffeine intake, dietary habits, screen time, alcohol and substance use, stress levels, and study habits all play a huge role in shaping sleep patterns. For instance, the CDC notes that inconsistent sleep patterns, common among college students, contribute to issues like “daytime sleepiness,” which can impair focus and academic success.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Our project seeks to understand which of these lifestyle habits most strongly affect sleep consistency and quality, focusing on students with demanding college schedules. To do so, we will examine measurable sleep factors like total sleep hours, consistency in average bedtime and wake time, and overall sleep quality. By analyzing these metrics, we aim to pinpoint lifestyle behaviors that correlate with healthier sleep patterns among college students in the U.S., hoping to provide practical insights into developing better sleep routines.

Previous research has highlighted key lifestyle and mental health factors that impact sleep consistency among U.S. college students. For instance, one study found that habits like caffeine use, substance use, and high stress are strongly associated with sleep issues, showing how these daily choices affect students’ ability to keep a regular sleep schedule.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Similarly, research on sleep hygiene—such as managing screen time before bed, limiting caffeine intake, and moderation of alcohol use—demonstrates how these habits impact sleep quality, often making it difficult for students to maintain a steady routine.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Another recent study on college students' sleep trends highlights that irregular schedules and high-stress levels are key contributors to poor sleep quality, emphasizing the importance of exploring how lifestyle and mental health impact sleep patterns.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

### EDA of Dataset #2

The primary goal of this dataset is to explore the sleep behaviors of university students and analyze how various lifestyle factors, such as screen time, caffeine intake, and physical activity, impact sleep quality and duration. By examining these relationships, the dataset provides insights into the influence of demographic factors like age, gender, and academic year on sleep patterns. Ultimately, the dataset aims to generate actionable recommendations for improving student wellness by addressing key areas of sleep health and lifestyle habits.

To ensure the dataset was ready for analysis, several data cleaning steps were undertaken. Rows with missing values were identified and removed to maintain data integrity and avoid analytical errors. Categorical variables, such as Gender and University Year, were standardized for consistency by capitalizing text. Sleep start and end times were converted into a 24-hour format to enable seamless analysis of temporal patterns. Additionally non-essential columns that did not contribute directly to the research objectives were excluded, allowing the focus to remain on variables directly related to sleep quality, duration, and lifestyle factors. These steps created a clean and structured dataset, facilitating meaningful exploratory and predictive analyses. 

```python
# converting convert decimal hours to HH:MM format
def decimal_to_time(decimal_hour):
    hours = int(decimal_hour)
    minutes = int((decimal_hour - hours) * 60)
    return f"{hours:02d}:{minutes:02d}"

time_columns = ['Weekday_Sleep_Start', 'Weekend_Sleep_Start', 'Weekday_Sleep_End', 'Weekend_Sleep_End']
for col in time_columns:
    sleep_data[col] = sleep_data[col].apply(lambda x: decimal_to_time(x) if pd.notnull(x) else x)
sleep_data.head()
```

```python
# handle missing values
missing_data = sleep_data.isnull().sum()
print("Missing values in each column:\n", missing_data)
```

```python
# drop rows with missing values
sleep_data.dropna(inplace=True)
```

```python
# clean categorical variables
sleep_data['Gender'] = sleep_data['Gender'].str.capitalize()
sleep_data['University_Year'] = sleep_data['University_Year'].str.lower()
```

```python
# display unique values in categorical columns
print('Unique values in Gender:', sleep_data['Gender'].unique())
print('Unique values in Univeristy Year:', sleep_data['University_Year'].unique())
```

```python
# summary stats to verify cleaning
sleep_data.describe()
```

```python
# Section 1: Overview of sleep patterns
plt.figure(figsize=(10, 6))
sns.histplot(data=sleep_data, x='Sleep_Duration', kde=True, color='skyblue', bins=20)
plt.title("Distribution of Sleep Duration")
plt.xlabel("Hours of Sleep")
plt.ylabel("Frequency")
plt.show()
```

This histogram shows the distribution of sleep duration, with most individuals sleeping between 5 to 7 hours. The x-axis represents hours of sleep, and the y-axis shows frequency, with a KDE curve highlighting a peak around 6 hours. This indicates that 6 hours is the most common sleep duration in the dataset, providing a baseline for analyzing sleep patterns further.

```python
#section 2: sleep duration and university year
order = ['1st year', '2nd year', '3rd year', '4th year']
sleep_data['University_Year'] = pd.Categorical(sleep_data['University_Year'], categories=order, ordered=True)
plt.figure(figsize=(10,6))
sns.boxplot(data=sleep_data, x='University_Year', y='Sleep_Duration', color='lightblue')
means = sleep_data.groupby('University_Year', observed=False)['Sleep_Duration'].mean()
for i, mean in enumerate(means):
    plt.scatter(i, mean, color='red', s=50, zorder=5, label='Mean' if i == 0 else "")
plt.title('Sleep Duration by University Year')
plt.xlabel('University Year')
plt.ylabel('Hours of Sleep')
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)
plt.show()
```

In this graph, we compared the sleep duration of students across university years, providing insights into how sleep patterns vary. The line in each box represents the **median sleep duration**, which remains fairly consistent at around 6-7 hours for all years. The height of the boxes and whiskers represents the variability in sleep duration. 1st year has the least variability, while 3rd year shows the most consistent sleep patterns. 4th year exhibits the widest range of sleep duration, with some students sleeping as little as 4 hours and others sleeping close to 9 hours. Overall, the plot suggests that sleep duration patterns are relatively stable across university years as the data is relatively well-distributed and no significant outliers are present.

```python
#Section 3: sleep quality of gender
plt.figure(figsize=(8, 6))
mean = sleep_data.groupby('Gender')['Sleep_Quality'].mean()
sns.boxplot(data=sleep_data, x='Gender', y='Sleep_Quality', color='lightblue')
for i, mean in enumerate(mean):
    plt.scatter(i, mean, color='red', s=50, zorder=5, label='Mean' if i == 0 else "")
plt.title("Sleep Quality Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Sleep Quality")
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)
plt.show()
```

In this boxplot, we analyze the distribution of sleep quality across three gender categories: Male, Female, and Other. The box represents the interquartile range (middle 50% of values), with the line inside the box indicating the median sleep quality. The red dots represent the mean sleep quality for each gender, and the whiskers extend to show the variability in sleep quality, excluding outliers.

For males, the sleep quality is centered around similar median and mean values with a moderate spread. Females exhibit slightly higher variability, with similar median and mean values. Individuals categorized as "Other" show slightly higher variability in sleep quality compared to males and females but maintain comparable median and mean values. Overall, the plot indicates that sleep quality distributions are consistent across genders, with no significant deviations observed.

```python
#section 4: impact of caffeine intake on sleep quality
heatmap_data = sleep_data.pivot_table(values='Sleep_Quality', index='Caffeine_Intake', columns='Gender', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".1f", cbar_kws={'label': 'Average Sleep Quality'})
plt.title('Heatmap of Average Sleep Quality by Caffeine Intake and Gender')
plt.xlabel('Gender')
plt.ylabel('Caffeine Intake (cups/day)')
plt.show()
```

This heatmap illustrates the relationship between caffeine intake (rows) and gender (columns) on average sleep quality. Warmer tones (red) indicate higher sleep quality, while cooler tones (blue) represent lower sleep quality. For example, females with no caffeine intake (0 cups/day) report the highest sleep quality (6.4), whereas higher caffeine levels are linked to a noticeable decline, particularly for females. Males, in contrast, show more stable sleep quality across all caffeine levels. The 'Other' group displays variability, suggesting additional factors may influence their sleep quality beyond caffeine intake.

```python
#section 5:
sleep_data['Screen_Time'] = pd.to_numeric(sleep_data['Screen_Time'], errors='coerce')
def classify_screen_time(hours):
    if hours <= 2.0:
        return 'Low'
    elif 2.0 < hours <= 3.5:
        return 'Moderate'
    else:
        return 'High'

sleep_data['Screen_Time'] = sleep_data['Screen_Time'].apply(classify_screen_time)
plt.figure(figsize=(10, 6))
sns.barplot(data=sleep_data, x='Screen_Time', y='Sleep_Quality', hue='Gender', errorbar=None, palette='viridis')
plt.title('Average Sleep Quality by Screen Time and Gender')
plt.xlabel('Screen Time')
plt.ylabel('Average Sleep Quality')
plt.legend(title='Gender', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

For this dataset, we define **consistency** as the variability in bedtimes and wake times across weekdays and weekends, measured in minutes. We categorized consistency into three levels:

- **High Consistency:** Variability less than or equal to 60 minutes.  
- **Moderate Consistency:** Variability between 61 and 120 minutes.  
- **Low Consistency:** Variability greater than 120 minutes.  

The following boxplots explore how consistent bedtimes and wake times (both on weekdays and weekends) affect sleep quality for males, females, and others.

```python
#section 6: Impact of Sleep Consistency on Sleep Quality by Gender
def categorize_consistency(minutes):
    if minutes <= 60:
        return "High Consistency"
    elif 61 <= minutes <= 120:
        return "Moderate Consistency"
    else:
        return "Low Consistency"

sleep_data['Weekday_Consistency'] = (
    sleep_data['Weekday_Sleep_Start'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1])) -
    sleep_data['Weekday_Sleep_End'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1])).abs()
)
sleep_data['Weekend_Consistency'] = (
    sleep_data['Weekend_Sleep_Start'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1])) -
    sleep_data['Weekend_Sleep_End'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1])).abs()
)

sleep_data['Weekday_Consistency_Level'] = sleep_data['Weekday_Consistency'].apply(categorize_consistency)
sleep_data['Weekend_Consistency_Level'] = sleep_data['Weekend_Consistency'].apply(categorize_consistency)

consistency_data = pd.melt(
    sleep_data,
    id_vars=['Gender', 'Sleep_Quality'],
    value_vars=['Weekday_Consistency_Level', 'Weekend_Consistency_Level'],
    var_name='Day_Type',
    value_name='Consistency_Level'
)

consistency_order = ['Low Consistency', 'Moderate Consistency', 'High Consistency']

consistency_data['Consistency_Level'] = pd.Categorical(
    consistency_data['Consistency_Level'],
    categories=consistency_order,
    ordered=True
)

genders = ['Male', 'Female', 'Other']
for gender in genders:
    plt.figure(figsize=(10, 6))
    sns.boxplot(
        data=consistency_data[consistency_data['Gender'] == gender],
        x='Consistency_Level',
        y='Sleep_Quality',
        hue='Day_Type',
        palette='viridis'
    )
    plt.title(f'Sleep Quality vs. Consistency Level for {gender}')
    plt.xlabel('Consistency Level')
    plt.ylabel('Sleep Quality')
    plt.legend(title='Day Type', loc='upper left', bbox_to_anchor=(1.05, 1))
    plt.show()
```

Based on these plots we can tell say tha for males, maintaining a consistent bedtime and wake time, with less than an hour of variability, is linked to overall better sleep quality. However, when their sleep schedules are inconsistent, particularly on weekends, their sleep quality tends to become more unpredictable and lower on average. Additionally, males experience slightly better sleep quality with consistent weekday schedules compared to weekends. 

For females, high consistency also improves sleep quality, though the impact is not as pronounced as it is for males. On the other hand, low consistency, especially during weekends, results in lower and more variable sleep quality. Moderate consistency in sleep schedules shows similar patterns for both weekdays and weekends for females. 

For individuals categorized as "Other," high consistency similarly contributes to better sleep quality, though some variation still exists. Notably, weekend consistency is generally associated with lower sleep quality compared to weekdays, especially in cases of moderate or low consistency. Overall, the findings highlight the importance of maintaining a regular sleep schedule to enhance sleep quality across all groups.

# Ethics & Privacy

In our project, there are potential unintended consequences that could arise from our analysis. By focusing on specific measurable factors, we may inadvertently oversimplify the complexity of sleep behaviors among students. For instance, our findings could reinforce stereotypes or generalizations about students with irregular sleep schedules without fully considering underlying causes like socioeconomic pressures, mental health challenges, or cultural factors. Such oversights might lead to conclusions that are not inclusive or effective for all students.

Our datasets may exclude or mislabel students who have underlying health conditions such as insomnia, mental health issues, or chronic illnesses. These populations might be underrepresented, and their data could be disproportionately treated as outliers and removed before the datasets were made accessible to us. Additionally, we acknowledge that the anonymized nature of our datasets limits our ability to confirm that the data truly represents college students across the US, which could introduce regional or demographic biases.

Socioeconomic factors present another challenge. For example, students from lower-income backgrounds may have formed sleep habits during high school while managing jobs or other responsibilities, contributing to what might be labeled as "unhealthy" sleep schedules. Our project does not directly measure these socioeconomic factors, and we acknowledge that this could limit the depth of our analysis. Similarly, interpersonal relationships and cultural practices, such as international students managing time zone differences to stay connected with family, could also influence sleep patterns in ways not captured in our surveys.

Moreover, we do not explicitly address how habits may have changed post-pandemic (e.g., altered sleep patterns due to remote learning or shifting routines) or the impact of environmental factors such as shared living spaces, loud roommates, or limited access to comfortable sleeping arrangements. These factors could significantly influence sleep quality but are not accounted for in our analysis.

To mitigate privacy concerns, our datasets anonymize participants by assigning numbers to each observation. While this ensures no personal or identifying information is used, it also limits our ability to contextualize the data fully. Participants' self-reported survey responses may also contain inaccuracies due to social desirability bias or hesitancy to disclose substance or alcohol use, potentially skewing our results.

To mitigate these biases and unintended consequences, we will:
1. Acknowledge the limitations of our data and analysis, particularly the absence of socioeconomic and environmental context.    
2. Avoid broad generalizations, presenting findings as trends specific to the datasets.
3. Recommend future research to explore underrepresented populations and additional variables like socioeconomic status, mental health, and environmental factors.

By critically reflecting on these aspects, we aim to ensure our study is ethical, inclusive, and aware of its limitations.
